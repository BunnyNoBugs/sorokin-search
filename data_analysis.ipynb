# Посмотрим, сколько существительных из всех являются матом и неприличными словами (жопа, говно, дерьмо) (извините)

```python
s = 0
for i in lemmas.split(' '):
    if i == "хуй":
        s+= 1
    elif i == "блять":
        s+= 1
    elif i == "пиздец":
        s+= 1
    elif i == "сука":
        s+= 1
    elif i == "нахуй":
        s+= 1
    elif i == "ебаный" or i == "блядь":
        s+= 1
    elif i == "хуйло":
        s+= 1
    elif i == "жопа" or i == "говно" or i == "дерьмо":
        s+= 1
```

```python
n = 0

for i in pos_tags.split(' '):
    
    if i == 'S':
        n += 1
```

```python
n = n - s
n
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(15, 13), subplot_kw=dict(aspect="equal"))

data= [ s,n]

ingredients = ['мат и всякое грубое', 'не мат']


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} )".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
  
                                  textprops=dict(color="b"))
ax.legend(wedges, ingredients,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))


plt.setp(autotexts, size=13, weight="bold")
plt.title('сколько существительных из всех являются матом и неприличными словами (жопа, говно, дерьмо) ')
fig.savefig('nouns.png')
fig.show()
```

# Посмотрим на биграммы

Что встречается чаще? Глагол + существительное (но нет существительного до глагола) или существительное + глагол (но нет существительного после глагола)

```python
s = 0
verb_noun = 0
noun_verb = 0
pst = pos_tags.split(' ')
for p in pst:
    if pst[s] == 'V' and pst[s+1] == 'S'and pst[s-1] != 'S':
        verb_noun += 1
        s += 1
    elif pst[s] == 'V' and pst[s-1] == 'S' and pst[s+1] != 'S':
        noun_verb += 1
        s += 1
    else:
        s += 1
```

```python
labels = ['Глагол + существительное', 'Существительное + глагол']
sizes = [verb_noun, noun_verb]
colors = ['yellowgreen', 'gold']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
